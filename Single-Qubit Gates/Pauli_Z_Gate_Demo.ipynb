```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector, visualize_transition, plot_histogram
from qiskit import transpile
from qiskit_aer import AerSimulator
```

```python
# Create a quantum circuit with a single qubit
# The default initial state of qubit will be |0> or [1,0]
qc = QuantumCircuit(1)
```

```python
#Apply the Pauli X-gate on the qubit to make the input to y gate as |1>
qc.x(0)
```

```python
#Apply the Pauli z-gate on the |1> qubit
qc.z(0)
```

```python
#Draw the circuit
# qc.draw()
qc.draw('mpl')
```

```python
# Add measurements
qc.measure_all()

# Get the backend for the circuit
backend = AerSimulator()

# Transpile the quantum circuit
qc_compiled = transpile(qc, backend)

# Execute the circuit on the simulator
job_sim = backend.run(qc_compiled, shots=1024)  # Increased shots for better statistics

# Grab the results from the job
result_sim = job_sim.result()

# Get the counts
counts = result_sim.get_counts(qc_compiled)
print(counts)
plot_histogram(counts)
```

```python
#plot the result as a bloch sphere visualization
plot_bloch_multivector(qc_compiled)
```

```python
# visualize the output as an animation
visualize_transition(qc)
```
